```python
#Cleaning Texts - Training
import pandas as pd
import re

def clean_text(text):
    text = text.strip()                           #Remove leading/trailing whitespace
    text = text.lower()                           #convert to lowercase
    text = re.sub(r'[^\w\s]', '', text)           #Remove punctuation
    text = re.sub(r'\s+', ' ', text)              #Collapse multiple spaces
    return text

df['Gender'] = df['Gender'].apply(clean_text)
df['MaritalStatus'] = df['MaritalStatus'].apply(clean_text)
df['PartTimeFullTime'] = df['PartTimeFullTime'].apply(clean_text)

df.head()
```

```python
#Label encoding Body part column as Encoded Keyword - Training
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['EncodedKeyword'] = le.fit_transform(df['BodyPart'])
```

```python
# One hot encoding Gender, MaritalStatus, PartTimeFullTime - Training
df = pd.get_dummies(df,columns=['Gender'],drop_first=True)
df = pd.get_dummies(df,columns = ['MaritalStatus'],drop_first=True)
df = pd.get_dummies(df,columns = ['PartTimeFullTime'],drop_first=True)

df.head()
```

```python
# Training our ML model - XG Boost
X_train = df.drop('UltimateIncurredClaimCost', axis=1)
y_train = df['UltimateIncurredClaimCost']
```

```python
import numpy as np

# Check for infinite values in training model
print(np.isinf(X_train).sum().sum())
```

```python
X_train.replace([np.inf, -np.inf], np.nan, inplace=True)
```

```python
X_train.fillna(X_train.mean(), inplace=True)
```

```python
print(np.isinf(y_train).sum().sum())
```

```python
from xgboost import XGBRegressor
model = XGBRegressor(random_state=51)
model.fit(X_train, y_train)
```

```python
#Applying our ML model on our test
X_test = df1[X_train.columns]
y_pred = model.predict(X_test)
```

```python
# Converting predicted values into thousands
y_pred = np.expm1(y_pred)
```

```python
y_pred_series = pd.Series(y_pred, name='Predictedvalues')
```

```python
X_test = X_test.reset_index(drop=True)
y_pred_series = y_pred_series.reset_index(drop=True)
```

```python
df_results = pd.concat([X_test, y_pred_series], axis=1)
```

```python
# Extracting predicted values to a CSV file
df_results.to_csv('test_with_predictions.csv', index=False)
```
